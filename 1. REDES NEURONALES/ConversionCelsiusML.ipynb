```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22],dtype = float)
farenheit =np.array([-40, 14, 32, 46, 59, 72], dtype= float)
```

Opción con una neurona:

```python
ModeloRN = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 1, input_shape =[1])
])
```

Opción con más neuronas:

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
ModeloRN = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
ModeloRN.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = "mean_squared_error"
)
```

```python
print("Entrenando el modelo...")
hist = ModeloRN.fit(celsius, farenheit, epochs = 500, verbose= False)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Epoca")
plt.ylabel("Perdida")
plt.plot(hist.history["loss"])
```

```python
ModeloRN.predict([100.0])
```

```python
ModeloRN.weights
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Datos de entrada
celsius = np.array([-40, -10, 0, 8, 15, 22], dtype=float)
farenheit = np.array([-40, 14, 32, 46, 59, 72], dtype=float)

# Cambiar la forma de los datos para que tengan la forma correcta para la regresión lineal
celsius = celsius.reshape(-1, 1)
farenheit = farenheit.reshape(-1, 1)

# Crear un modelo de regresión lineal
ModeloLineal = LinearRegression()

# Entrenar el modelo con los datos
ModeloLineal.fit(celsius, farenheit)

# Imprimir los coeficientes del modelo
print("Coeficiente:", ModeloLineal.coef_)
print("Intercepto:", ModeloLineal.intercept_)
```

```python
ModeloLineal.predict(np.array([[100]]))
```
